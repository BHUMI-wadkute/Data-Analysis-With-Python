```python
data['reading_score'].corr(data['writing_score'])
```

```python
data['xd'] = (data['math_score'] + data['reading_score'] + data['writing_score'])
x = data['xd'].mean()
data = data[data['xd']>x]
p = data[data['parent_education'] == "Bachelor's"]
len(p)
```
