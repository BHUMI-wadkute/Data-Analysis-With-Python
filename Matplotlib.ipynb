```python
plt.figure(figsize = (8,6),dpi=100,)
plt.plot(df['Month'],df['Sales'], marker = 'o',ms=15, mfc='r', ls = 'dotted')
# plt.plot(df['Sales'])
# plt.plot(df['Profit'])
plt.title('my graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.savefig('plot.png')
plt.show()   #scatter #pie #bar
```

```python
plt.figure(figsize = (8,6))
plt.bar(df['Month'],df['Sales'], )
plt.title('my graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.savefig('plot.png')
plt.show()  
```
